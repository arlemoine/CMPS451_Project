```python
import matplotlib.pyplot as plt
import numpy as np
from scipy import stats
import statsmodels.api as sm
```

```python
# Datasets
rng = np.random.default_rng(seed=43)
dataset1_norm = rng.normal(loc=10, scale=2, size=50)
dataset2_uni = rng.uniform(low=4, high=16, size=50)

print(dataset1_norm)
print(dataset2_uni)
```

```python
# Sort the datasets
dataset1_sorted = np.sort(dataset1_norm)
dataset2_sorted = np.sort(dataset2_uni)

print(dataset1_sorted)
print(dataset2_sorted)
```

```python
# Establish quantiles
quantiles = [0.1, 0.25, 0.5, 0.75]
quantile_labels = ['10% Quantile', '25% Quantile', '50% Quantile', '75% Quantile']
dataset1_quantile = np.quantile(dataset1_sorted, quantiles)
dataset2_quantile = np.quantile(dataset2_sorted, quantiles)
```

```python
# Plotting
fig, ax = plt.subplots(figsize=(8, 6))

ax.set_title('Q-Q Plot: Normal vs. Uniform Datasets')
ax.set_xlabel('Quantiles of Uniform Data')
ax.set_ylabel('Quantiles of Normal Data')
ax.grid(True)

ax.scatter(dataset1_sorted, dataset2_sorted, c='gray', label='Data Quantiles')
colors = ['red', 'orange', 'yellow', 'green', 'blue', 'purple']
for i in range(len(quantiles)):
    ax.scatter(dataset1_quantile[i], dataset2_quantile[i], c=colors[i], label=quantile_labels[i])

# Find the overall minimum and maximum values from both datasets
min_val = min(dataset1_sorted.min(), dataset2_sorted.min())
max_val = max(dataset1_sorted.max(), dataset2_sorted.max())

print(min_val)
print(max_val)

# Plot the reference line??

ax.legend()
plt.show()
```
